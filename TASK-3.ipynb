```python
# importing opencv,numpy

import numpy as np
import cv2
from google.colab.patches import cv2_imshow
```

```python
#reading an image

image=cv2.imread('sundar.jfif')
cv2_imshow(image)
```

```python
# converting image to grayscale

gray_image=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_image)
```

```python
# inverting grayscale image

inverted_gray_image=255-gray_image
cv2_imshow(inverted_gray_image)
```

```python
# Blurring the inverted grayscale image

blur_image=cv2.GaussianBlur(inverted_gray_image,(5,5),3)
cv2_imshow(blur_image)
```

```python
# Inverting the blur image 

inverted_blur_image=255-blur_image
cv2_imshow(inverted_blur_image)
```

```python
# Dividing grayscale and inverted blur image to get sketch type structure

pencil_sketch=cv2.divide(gray_image,inverted_blur_image,scale=255.0)
#cv2_imshow(pencil_sketch)
```

```python
cv2.imwrite('original.jpg',image)
cv2.imwrite('sketch.jpg',pencil_sketch)
```

```python

print('Original Image \n')
cv2_imshow(image)
print('\n--------------------------------------------\n')
print('Pencil Sketch  \n')
cv2_imshow(pencil_sketch)
```
